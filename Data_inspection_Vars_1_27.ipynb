# Make lists of Categorical (descriptive, unordered), 
# Ordinal (ordered), and Numeric Variables

```python
cat_vars = ['MSSubClass', 'MSZoning', 'Street', 'Alley', 'LandContour', 'LotConfig', 'Neighborhood', 
            'Condition1', 'Condition2', 'BldgType', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd',
            'MasVnrType']

ordinal_vars = ['LotShape', 'Utilities', 'LandSlope', 'HouseStyle', 'ExterQual']
# 'OverallQual', 'OverallCond' are already numeric, running from 1-10

numeric_vars = ['LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea']
```

# Encoding 15 Categorical Variables

```python
#Encode Categotical Variables

encoder = OneHotEncoder(sparse_output=False, drop=None, handle_unknown='error')
encoded_cats = encoder.fit_transform(df_selected[cat_vars])

feature_names = encoder.get_feature_names_out(cat_vars)

# Create a new dataframe with encoded variables
encoded_df = pd.DataFrame(encoded_cats, columns=feature_names, index=df_selected.index).astype(bool)

# Drop and merge
df_selected = df_selected.drop(columns=cat_vars)
df_final = pd.concat([df_selected, encoded_df], axis=1)

# Verification prints
print("Original df_selected shape:", df_selected.shape)
print("encoded_df shape:", encoded_df.shape)
print("Final merged df shape:", df_final.shape)
print("\nColumn types in final dataframe:")
print(df_final.dtypes.value_counts())
```

```python
df_final.info()
```

# Recode Ordinal Variables

```python
# LotShape: General shape of property

print(df_final['LotShape'].unique()) 
print(df_final['LotShape'].value_counts()) 

# Map
LotShape_map = {
       'Reg': 4, #	Regular
       'IR1': 3, #	Slightly irregular
       'IR2': 2, #	Moderately Irregular	
       'IR3': 1, #	Irregular
}

df_final.loc[:, 'LotShape'] = df_final['LotShape'].map(LotShape_map)

# Convert to Int64
df_final = df_final.copy()
df_final['LotShape'] = df_final['LotShape'].astype('Int64')

# Check
print(df_final.info())
print("Final unique values:", df_final['LotShape'].unique())
print("Final unique counts:", df_final['LotShape'].value_counts())
print(df_final.shape)
```

```python
#Utilities: Type of utilities available
print(df_final['Utilities'].unique()) 
print(df_final['Utilities'].value_counts()) 

# Map
Utilities_map = {
       'AllPub': 4, #	All public Utilities (E,G,W,& S)	
       'NoSewr': 3, #	Electricity, Gas, and Water (Septic Tank)
       'NoSeWa': 2, #	Electricity and Gas Only	
       'ELO': 1,    #	Electricity only
}

df_final.loc[:, 'Utilities'] = df_final['Utilities'].map(Utilities_map)

# Convert to Int64
df_final = df_final.copy()
df_final['Utilities'] = df_final['Utilities'].astype('Int64')

# Check
print(df_final.info())
print("Final unique values:", df_final['Utilities'].unique())
print("Final unique counts:", df_final['Utilities'].value_counts())
print(df_final.shape)
```

```python
# LandSlope: Slope of property
print(df_final['LandSlope'].unique()) 
print(df_final['LandSlope'].value_counts()) 

# Map
LandSlope_map = {
       'Gtl': 3, #	Gentle Slope
       'Mod': 2, #	Moderate Slope
       'Sev': 1, #	Severe Slope
}

df_final.loc[:, 'LandSlope'] = df_final['LandSlope'].map(LandSlope_map)

# Convert to Int64
df_final = df_final.copy()
df_final['LandSlope'] = df_final['LandSlope'].astype('Int64')

# Check
print(df_final.info())
print("Final unique values:", df_final['LandSlope'].unique())
print("Final unique counts:", df_final['LandSlope'].value_counts())
print(df_final.shape)
```

```python
# HouseStyle: Style of dwelling
print(df_final['HouseStyle'].unique()) 
print(df_final['HouseStyle'].value_counts()) 

# Map
HouseStyle_map = {
       '1Story': 100, #	One story
       '1.5Fin': 155, #	One and one-half story: 2nd level finished
       '1.5Unf': 150, #	One and one-half story: 2nd level unfinished
       '2Story': 200, #	Two story
       '2.5Fin': 255, #	Two and one-half story: 2nd level finished
       '2.5Unf': 250, #	Two and one-half story: 2nd level unfinished
       'SFoyer': 301, #	Split Foyer
       'SLvl':	 300, # Split Level
}

df_final.loc[:, 'HouseStyle'] = df_final['HouseStyle'].map(HouseStyle_map)

# Convert to Int64
df_final = df_final.copy()
df_final['HouseStyle'] = df_final['HouseStyle'].astype('Int64')

# Check
print(df_final.info())
print("Final unique values:", df_final['HouseStyle'].unique())
print("Final unique counts:", df_final['HouseStyle'].value_counts())
print(df_final.shape)	
       
```

```python
df_final['HouseStyle_simple'] = df_final['HouseStyle']

def simplify_housestyle(value):
    if value in [100, 150, 155]:  # Single-level styles
        return 1
    elif value in [200, 250, 255]:  # 2-level styles
        return 2
    elif value in [300, 301 ]:  # split-level styles
        return 3
    else:
        return None 

df_final['HouseStyle_simple'] = df_final['HouseStyle'].apply(simplify_housestyle)

# Check the results
print(df_final['HouseStyle_simple'].unique())
print(df_final['HouseStyle_simple'].value_counts())
```

```python
# ExterQual: Evaluates the quality of the material on the exterior 
print(df_final['ExterQual'].unique()) 
print(df_final['ExterQual'].value_counts()) 

# Map
ExterQual_map = {
    'Ex': 5,  # Excellent
    'Gd': 4,  # Good
    'TA': 3,  # Average/Typical 
    'Fa': 2,  # Fair 
    'Po': 1,  # Poor 
}

df_final.loc[:, 'ExterQual'] = df_final['ExterQual'].map(ExterQual_map)

# Convert to Int64
df_final = df_final.copy()
df_final['ExterQual'] = df_final['ExterQual'].astype('Int64')

# Check
print(df_final.info())
print("Final unique values:", df_final['ExterQual'].unique())
print("Final unique counts:", df_final['ExterQual'].value_counts())
print(df_final.shape)	
```
